```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

#take all point in population
x = [54.4,102.4,52.2,122.9,66.6,47.2,49.1]
y = [2951.5,2361.9,1762.2,3049.6,1934.3,1984.1,1800.7]
z = [12,56,63,58,79,33,57]

x = np.array(x)
y = np.array(y)
z = np.array(z)

plt.rcParams['font.size'] = '18'
fig = plt.figure()
ax = fig.add_subplot(111)
fig.set_size_inches(18, 10)
plt.scatter(x,y, c = 'green')
plt.xlabel('Time (minutes)')
plt.ylabel('Cost (dollars)')
plt.title('Cost vs. Time')

for xy in zip (x,y):
    ax.annotate('(%s,%s)' %xy, xy=xy, textcoords='data',
                ha='left',va='center')

filename =  'Cost-Vs-Time'  +'.png'
plt.savefig(filename)
plt.show()


fig = plt.figure()
fig.set_size_inches(18, 10)
ax = fig.add_subplot(111)
plt.scatter(y,z, c = 'blue')

plt.xlabel('Cost (dollars)')
plt.ylabel('Quality (100 point scale)')
plt.title('Quality vs. Cost')

for yz in zip (y, z):
    ax.annotate('(%s,%s)' %yz, xy=yz, textcoords='data',
                ha='left',va='center')

filename =  'Quality-Vs-Cost'  +'.png'
plt.savefig(filename)
plt.show()
```
